## 8. GRAPHS

```python
import plotly.express as px
```

## 8.1 Line graph

```python
Usme = n2.query("area == 'Usme'")
Usme.head()
```

```python
# Fig 1
fig1 = px.line(Usme, x = 'año_str', y = 'cep', title = "Cases of excess weight 2007-2023")
fig1.show()
```

```python
# Fig 2
fig2 = px.line(Usme, x = 'año_str', y = 'cep')

#title
fig2.update_layout(title_text = "Cases of excess weight 2007-2023", title_font_size = 18, title_x = 0.5)


fig2.show()

#axis
fig2.update_xaxes(title = "year", title_font_size = 18)
fig2.update_yaxes(title = "Cases of excess weight 2007-2023", title_font_size = 18)

fig2.show()
```

```python
# Fig 3
fig3 = px.bar(Usme, x = 'año_str', y = 'cep')

# title

fig3.update_layout(title_text = 'Cases of excess weight in Usme',
                   title_font_size = 30,
                   title_x = 0.5)

# axis
fig3.update_xaxes(title = 'Years', title_font_size = 20)
fig3.update_yaxes(title = 'number of cases', title_font_size = 20)

fig3.show()
```

## **INTERPRETATION FIG 3**

The graph shows the number of cases of excess weight in the town of Usme from 2007 to 2023. The y-axis represents the number of cases, while the x-axis represents the year. The line on the graph shows a general upward trend, which means that the number of cases of excess weight has increased steadily over the years. This trend is worrying, as excess weight is a risk factor for a number of serious health problems, including diabetes, heart disease and some types of cancer.It is important to note that the graph only shows the number of diagnosed cases of excess weight. There are likely many more undiagnosed cases. Additionally, the graph does not take into account other factors that may contribute to excess weight.However, the graph provides a worrying picture of the current trend of excess weight in Usme. It is important to take steps to prevent excess weight and promote a healthy lifestyle.

## 8.3 More than one line

```python
## Create a list of countries
c_list = ['Usaquén', 'Chapinero', 'Santa fe', 'San Cristóbal', 'Usme', 'Ciudad Bolívar']

## Extract the countries from the dataframe
### First we must unindexed the df
n3 = n2.reset_index()
area = n3[n3['area'].isin(c_list)]

area.head()
```

```python
fig4 = px.line(area, x = 'año_str', y = 'crt', color = 'area')

# title

fig4.update_layout(title_text = 'Cases of delay in size in localities',
                   title_font_size = 30,
                   title_x = 0.5)

# axis
fig4.update_xaxes(title = 'Years', title_font_size = 20)
fig4.update_yaxes(title = 'Number of cases', title_font_size = 20)

fig4.show()
```

## **INTERPRETATION FIG 4**


The graph shows that the number of cases of stunting in locations in Bogota has increased significantly in the last 16 years. This trend is worrying, since stunting is a serious health problem that can have long-term consequences for a child's physical and cognitive development.The graph also shows that the number of cases of delayed height in localities varies depending on the locality. For example, in 2023, there were approximately 926 cases in Usaquén, while there were approximately 1,791 cases in Ciudad Bolívar. This variation may be due to a number of factors, such as socioeconomic status, access to healthcare, and diet quality.

It is important that health authorities investigate the causes of the variation in the number of stunting cases in localities and take measures to ensure that all children have access to the care and resources they need to grow and develop healthily.

## 8.4 Pie Chart


```python
# The idea is to graph the top 20 countries
# by population and the year 2020

p2021_2 = p2021.sort_values('cd', ascending = False)
p2021_2.head()

## Top 20 by population
p2021_2 = p2021_2.iloc[0:20]
```

```python
fig5 = px.pie(p2021_2, values = 'cd', names = p2021_2.index.get_level_values(0))

# title
fig5.update_layout(title_text = 'cases of thinness in 20 locations in Bogota',
                   title_font_size = 30,
                   title_x = 0.5)
fig5.show()
```

## **INTERPRETATION FIG 5**

The graph shows the distribution of cases of thinness in the 20 locations of Bogotá, Colombia. The graph is presented in the form of a pie chart, where each sector represents a locality and the size of the sector is proportional to the number of cases of thinness recorded in that locality.
The town with the highest number of cases of thinness is Kennedy, with 15.8% of the total.
They are followed by Suba, with 14.7%, and Engativá, with 1.33%.
The towns with the lowest number of cases of thinness are Bosa, with 1.46%, los Martires 1.33% and Antoni Nariño 0.94%; It also shows that the distribution of cases of thinness in the 20 localities of Bogotá is unequal. Some towns, such as Kennedy, Suba and Engativá, have a significantly higher concentration of cases of thinness than others.

This information can be useful for health authorities to identify localities that require greater attention and to develop prevention and care strategies for thinness.
