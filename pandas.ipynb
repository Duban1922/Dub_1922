## CLEANING

```python
n.rename(columns= {"Año":"año"}, inplace = True)
n.rename(columns= {"cod":"cod"}, inplace = True)
n.rename(columns= {"Área":"area"}, inplace = True)
n.rename(columns= {"Casos delgadez":"cd"}, inplace = True)
n.rename(columns= {"Proporción delgadez":"pd"}, inplace = True)
n.rename(columns= {"Casos exceso de peso":"cep"}, inplace = True)
n.rename(columns= {"Proporción exceso de peso":"pep"}, inplace = True)
n.rename(columns= {"Casos Retraso en talla ":"crt"}, inplace = True)
n.rename(columns= {"Proporción Retraso en talla ":"prt"}, inplace = True)
n.rename(columns= {"Total Poblacion":"tp"}, inplace = True)


```

```python
n2 = n
```

```python
n.head()
```

##DELETE COLUMNS

```python
n2 = n2.drop(columns= ["Cod"])
```

```python
n2.head()
```

```python
n2["año_str"] = n2["año"].astype(str)
```

# **5: Indexing**



```python
n2 = n2.set_index(["area", "año"])

n2.head()
```

## 8. GRAPHS

```python
import plotly.express as px
```

## 8.1 Line graph

```python
Usme = n2.query("area == 'Usme'")
Usme.head()
```

```python
# Fig 1
fig1 = px.line(Usme, x = 'año_str', y = 'cep', title = "Casos de exceso de peso 2008-2023")
fig1.show()
```

```python
# Fig 2
fig2 = px.line(Usme, x = 'año_str', y = 'cep')

#title
fig2.update_layout(title_text = "Casos de exceso de peso 2008-2023", title_font_size = 18, title_x = 0.5)


fig2.show()

#axis
fig2.update_xaxes(title = "año", title_font_size = 18)
fig2.update_yaxes(title = "Casos de exceso de peso", title_font_size = 18)

fig2.show()
```
